```python
country_names = final_df.Country.value_counts().index
```

```python
country_values = final_df.Country.value_counts().values
```

```python
# Pie chart- top 3 countries
plt.pie(country_values[:3],labels= country_names[:3],autopct='%1.2f%%')
```

Observation: zomato maximum records or transactions are from India
after that USA and then UK.

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

## observations:
#### when rating between

1. 4.5 to 4.0 ------>excellent
2. 4.0 to 3.4----->very good
3. 3.5 to 3.9-----> good
4. 3.0 to 3.5------> avg
5. 2.5 to 2.9-----> average
6. 2.0 to 2.4 -------> poor
7. 0.1 to 1.9--------> poor 
8. 0 ------> not rated

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(data=ratings, x='Aggregate rating', y='Rating Count')
```

```python
sns.barplot(data=ratings, x='Aggregate rating', y='Rating Count', hue= 'Rating color',palette=['blue','red','orange','yellow','green','green'])
```

observation:
1. not rated count is very high
2. max no of rating is between 2.5 to 3.4

```python
## count plot
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
