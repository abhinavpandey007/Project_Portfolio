### feature enginnering process

```python
#splitting  DOJ feature in three diff categories
final_df['Date'] = final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month'] = final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year'] = final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head(2)
```

```python
# types casting the date,year and month features to int---->
final_df.Date = final_df.Date.astype(int)
final_df.Month = final_df.Month.astype(int)
final_df.Year = final_df.Year.astype(int)
```

```python
final_df.info()
```

```python
final_df['Year'].unique()
```

```python
#dropping DOJ since we splitted that feature...
#dropping year feature , since 2019 is unique value for all records & data is from 2019 only
final_df.drop('Date_of_Journey',axis=1,inplace=True)
final_df.drop('Year',axis=1,inplace=True)
```

```python
# dropping duplicates if any
final_df.duplicated().sum()
```

```python
final_df.drop_duplicates(keep='first',inplace=True)
#dropped duplicates keeping first records(246 duplicate values)
```

```python
# splitting arrival_time with time of arrival only removing date and month from that feature
final_df['Arrival_Time'] = final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
final_df
```

```python
#splitting arrival_time in hours and mints
final_df['Arrival_hour'] = final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_min'] = final_df['Arrival_Time'].str.split(':').str[1]
```

```python
final_df.head()
```

```python
#removing arrival_time feature since we made hours and min time feature out of that
final_df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
#convert hours and min to int type from object dtype
final_df['Arrival_hour'] = final_df['Arrival_hour'].astype(int)
final_df['Arrival_min'] = final_df['Arrival_min'].astype(int)
```

```python
final_df.info()
```

```python
#splitting dep_time in hours and mints
final_df['Dept_hour'] = final_df['Dep_Time'].str.split(':').str[0]
final_df['Dept_min'] = final_df['Dep_Time'].str.split(':').str[1]
```

```python
#dropping dep_time
final_df.drop('Dep_Time',axis=1,inplace=True)
```

```python
#converting dep_hour and min to int dtype from object dtype.
final_df['Dept_hour'] = final_df['Dept_hour'].astype(int)
final_df['Dept_min'] = final_df['Dept_min'].astype(int)
```

```python
#splitting duration in hours and mints
final_df['duration_hour'] = final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
final_df['duration_min'] = final_df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
#dropping 5m reocrds in duration_hour
final_df.drop([6474,2660],axis=0,inplace=True)
```

```python
#replacing nan values with zero in duration_min since we need to convert that object to int
final_df.duration_min.replace(np.nan,0,inplace=True)
```

```python
#converting dtypes of duration_hours_min to int & droppping duration hours--->
final_df.duration_hour = final_df.duration_hour.astype(int)
final_df.duration_min = final_df.duration_min.astype(int)

final_df.drop('Duration',axis=1,inplace=True)
```

```python
# Normalizing train(X) data
def norm_func(i):
    x = (i-i.min())/ (i.max()-i.min())
    return (x)
```

```python
# Normalized data frame (considering the encoded and train data)
X_n = norm_func(X.iloc[:, :])
X_n.describe()
```

# Using KNN algorithm


```python
#importing knn from sklearn

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 11)
```

```python
KNN = knn.fit(X_train, Y_train)  # Train the kNN model
```

### Evaluate Model

```python
#evaluate the model with train data

pred_train = knn.predict(X_train)  # Predict on train data
pred_train
```

```python
#cross table, for confusion matrices
pd.crosstab(Y_train, pred_train, rownames = ['Actual'], colnames = ['Predictions'])
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_train, pred_train))  #accuracy measure
```

#### got accuracy of 80%, but need more accuracy so


```python
#predict the class on test data

pred = knn.predict(X_test)
pred
```

```python
#evaluate the model with test data

print(accuracy_score(Y_test, pred))
pd.crosstab(Y_test, pred, rownames = ['Actual'], colnames= ['Predictions'])
```

```python
#creating empty list variable 
acc = []

#running KNN algorithm for 3 to 50 nearest neighbours(odd numbers) and 
#storing the accuracy values
for i in range(3, 50, 2):
    neigh = KNeighborsClassifier(n_neighbors=i)
    neigh.fit(X_train, Y_train)
    train_acc = np.mean(neigh.predict(X_train) == Y_train)
    test_acc = np.mean(neigh.predict(X_test) == Y_test)
    diff = train_acc - test_acc
    acc.append([diff, train_acc, test_acc,i])
```
